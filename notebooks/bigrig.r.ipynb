```r
library('ggplot2')

options(repr.plot.width =15, repr.plot.height = 11) 

dir.create("figs")
```

```r
# start coding here
df = read.csv('bigrig_times.csv')

df$franges <- as.factor(df$ranges)
df$ftaxa <- as.factor(df$taxa)
df$froot.range <- as.factor(df$root.range)


df
```

```r
ggplot(data=df, aes(x=ftaxa, y=time, fill=franges)) + 
    geom_boxplot(width=0.9) + 
    scale_color_brewer(palette="Pastel1") + 
    xlab("Taxa") + 
    ylab("Time (s)") +
    labs(fill = "Range") +
    scale_y_continuous(trans='log2', breaks = scales::pretty_breaks(7))+
    theme_bw() +
    theme(axis.text.x = element_text(angle = 45,  margin=margin(12,2,2,2,"pt")), axis.text.y = element_text(angle = 45))
```

```r
ggsave("figs/bigrig.times.boxplot.svg", width=15, height=11)
```

```r
ggplot(data=df, aes(x=taxa, y=time, colour=franges)) +
    geom_smooth() +
    scale_color_brewer(palette="Pastel1") + 
    xlab("Taxa") +
    ylab("Time (s)") +
    labs(colour="Range") + 
    scale_y_continuous(trans='log2', breaks = scales::pretty_breaks(n = 7))+
    scale_x_continuous(trans='log2', breaks = unique(df$taxa))+
    guides(x = "axis_logticks", y = "axis_logticks") +
    theme_bw() +
    theme(axis.text.x = element_text(angle = 45,  margin=margin(12,2,2,2,"pt")), axis.text.y = element_text(angle = 45))
```

```r
ggsave("figs/bigrig.times.linreg.svg", width=15, height=11)
```
