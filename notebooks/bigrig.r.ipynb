```r
library('ggplot2')

options(repr.plot.width = 15, repr.plot.height =11) 
```

```r
# start coding here
df = read.csv('bigrig_times.csv')

df$franges <- as.factor(df$ranges)
df$ftaxa <- as.factor(df$taxa)
df$froot.range <- as.factor(df$root.range)


df
```

```r
ggplot(data=df, aes(x=franges, y=time, fill=ftaxa)) + 
    geom_boxplot(width=0.2) + 
    scale_color_brewer(palette="Pastel1") + 
    xlab("Range") + 
    ylab("Time") +
    labs(fill = "Taxa") +
    theme_bw()
```

```r
ggplot(data=df, aes(x=taxa, y=time, colour=franges)) +
    geom_smooth() +
    scale_color_brewer(palette="Pastel1") + 
    xlab("Taxa") +
    ylab("Time (s)") +
    labs(colour="Range") + 
    theme_bw()
```
