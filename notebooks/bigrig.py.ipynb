```python
import pandas
import seaborn

seaborn.set_theme(style="ticks", palette="pastel", rc={'figure.figsize':(11.7,8.27)})

df = pandas.read_csv('bigrig_times.csv')
df
```

```python
g = seaborn.boxplot(df, x = "ranges", y="time", hue="taxa", palette = "pastel", width = 0.1)
g.label(y="Time", x="Ranges")
```
